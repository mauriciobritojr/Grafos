#### Importa bibliotecas necessárias

```python
import networkx as nx
from networkx.algorithms import bipartite
from networkx.drawing.layout import bipartite_layout
import matplotlib.pyplot as plt
```

#### Ler arquivo txt para gerar vértices e arestas

```python
arquivo = open("grafo_teste_1.txt", "r")
arestas = []
lista = []
for linha in arquivo:
    linha_limpa = linha.rstrip().split(' ')
    lista = lista + linha_limpa
    arestas.append(tuple(linha_limpa))
vertices = set(lista)
print(f'Total de Vértices: {len(vertices)} | Total de Arestas: {len(arestas)}')
```

#### Gera vértices e arestas do grafo G

```python
G = nx.Graph()
G.add_nodes_from(vertices)
G.add_edges_from(arestas)
```

#### Mostra o grafo G

```python
fig = plt.figure(figsize=(10,10))
nx.draw(G, pos=nx.spring_layout(G), with_labels=False, node_color='red', node_size=2, edge_color='orange', width=0.3)
plt.show()
```

#### Verifica se o grafo é bipartido

```python
print(bipartite.is_bipartite(G))
```

#### Separa os Vértices do Grafo em Dois Conjuntos

```python
bottom_nodes, top_nodes = list(bipartite.sets(G))

color_map = []
for v in bottom_nodes:
    color_map.append('red')
for v in top_nodes:
    color_map.append('blue')
```

```python
fig = plt.figure(figsize=(10,10))
nx.draw(G, pos=nx.spring_layout(G), with_labels=False, node_color=color_map, node_size=3, edge_color='orange', width=0.2)
plt.show()
```

#### Layout Bipartido

```python
# G = nx.bipartite.gnmk_random_graph(3, 5, 10, seed=123)
top = nx.bipartite.sets(G)[0]
pos = nx.bipartite_layout(G, top, )
```

```python
fig = plt.figure(figsize=(10,10))
nx.draw(G, pos=pos, with_labels=False, node_color=color_map, node_size=3, edge_color='orange', width=0.2)
plt.show()
```
