### Grafos - Exercícios Aula 09/03/21
Questão: Em um canil de reabilitação de cães abandonados, a equipe acabou de receber 6 novos cachorros e precisa colocá-los em diferentes casinhas para passar a primeira noite. Constatou-se que alguns desses cães ficam extremamente agressivos na presença de alguns do mesmo grupo, logo terão que alocá-los em casinhas diferentes:
- O cachorro A não pode ficar com: C, D ou E;
- Os cachorros B e F não podem ficar juntos;
- O cachorro E não pode ficar junto de D ou F.
##### Quantas casinhas serão necessárias para acomodar os 6 cachorros?
  
O número cromático será equivalente ao mínimo número de casinhas para acomadr os cães na primeira noite.

#### Definição (Coloração)
Seja G(V, A) um grafo e C = {c1, c2, c3, . . . , cm} um conjunto de cores. Uma coloração de G é uma atribuição  
de cores aos vértices de G de tal forma que dois vértices adjacentes recebam cores diferentes. O número cromático  
de um grafo G é o menor número de cores necessário para obter uma coloração de G. Se o número cromático é χ(G),  
dizemos que o grafo é χ(G)-cromático.

```python
import networkx as nx
```

### Função para determinar o número cromático de um grafo e associar cores distintas a cada vértice.

```python
def colore_vertices(G):
    paleta = {0:'blue', 1:"Green", 2:"red", 3:"orange", 4:"violet"}
    color_map = []
    # Determina o número cromático do Grafo, ou seja, a menor quantidade de cores que colore o grafo conforme as restrições
    cor_vertices = nx.coloring.greedy_color(G, strategy="largest_first")
    nr_cromatico = max(cor_vertices, key=cor_vertices.get)
    print("Número mínimo de Cores:", cor_vertices[nr_cromatico]+1)
    
    # define a cor de cada vértice
    for node in G:
        color_map.append(paleta[cor_vertices[node]])
    return color_map
```

```python
# Definição do grafo
G = nx.Graph()

V = ["A", "B", "C", "D", "E", "F"]
E = [("A", "C"), ("A", "D"), ("A", "E"), ("B", "F"), ("E", "D"), ("E", "F")]
G.add_nodes_from(V)
G.add_edges_from(E)

nx.draw(G, pos=nx.spring_layout(G), with_labels=True, font_weight='bold', font_color="w", node_color='b')
```

```python
# Mostra o grafo colorido
color_map = colore_vertices(G)
nx.draw(G, pos=nx.spring_layout(G), with_labels=True, font_weight='bold', font_color="w", node_color=color_map)
```
