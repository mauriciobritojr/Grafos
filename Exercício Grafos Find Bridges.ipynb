Questão: Dado o grafo abaixo, identifique as pontes deste (se houver).
 
Algoritmo:  
Para todas as arestas (u,v), faça:  
  1. Remove (u,v) do grafo;
  2. Se o grafo continuar conectado
          Então: (u,v) não é ponte;
          Senão: (u,v) é ponte;
  3. Adiciona (u,v) ao grafo;

```python
import networkx as nx
```

```python
# Find Bridges
def find_bridges(G):
    bridges = []
    colors = nx.get_edge_attributes(G,'color').values()
    for edge in E:
        G.remove_edge(edge[0],edge[1])
        if nx.is_connected(G):
            G.add_edge(edge[0],edge[1],color='k')
        else:
            bridges.append(edge)
            G.add_edge(edge[0],edge[1],color='r')
    colors = nx.get_edge_attributes(G,'color').values()
    nx.draw(G, pos=nx.spring_layout(G), with_labels=True, font_weight='bold', font_color="w", node_color='b', edge_color=colors)
    print(f'Total de Pontes de G:{len(bridges)}, Pontes:{bridges}')
```

```python
# Definição do grafo
G = nx.Graph()

V = [1, 2, 3, 4, 5, 6, 7, 8, 9]
E = [(1,2), (1,4), (2,3), (3,4), (4,5), (5,6), (5,8), (5,9), (6,7), (6,8), (8,9)]
G.add_nodes_from(V)
G.add_edges_from(E)

nx.draw(G, pos=nx.spring_layout(G), with_labels=True, font_weight='bold', font_color="w", node_color='b')
```

```python
find_bridges(G)
```

```python
# Adiciona vértices em um loop
G = nx.Graph()
for v in range(10):
    G.add_node(v, color='k')
E = [(0,3), (1,8), (2,8), (2,0), (2,3), (3,8), (8,10), (4,10), (4,7), (7,10), (4,9), (5,9), (5,6), (9,6)]
G.add_edges_from(E)
nx.draw(G, pos=nx.spring_layout(G), with_labels=True, font_weight='bold', font_color="w", node_color='b')
```

```python
find_bridges(G)
```
