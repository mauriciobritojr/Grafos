### Grafos - Exercícios Aula 09/03/21
Questão: Em um Pet Shop, uma vendedora precisa expor 7 peixes exóticos em alguns aquários para venda. Alguns desses peixes não podem ficar juntos por risco de conflito. A  tabela abaixo indica com "x" os peixes que não podem ficar juntos:

##### Qual o menor número de aquários necessários para expor os peixes sem conflito?
  
O número cromático será equivalente ao mínimo número de aquários necessários para expor os peixes.

#### Definição (Coloração)
Seja G(V, A) um grafo e C = {c1, c2, c3, . . . , cm} um conjunto de cores. Uma coloração de G é uma atribuição  
de cores aos vértices de G de tal forma que dois vértices adjacentes recebam cores diferentes. O número cromático  
de um grafo G é o menor número de cores necessário para obter uma coloração de G. Se o número cromático é χ(G),  
dizemos que o grafo é χ(G)-cromático.

```python
import networkx as nx
```

### Função para determinar o número cromático de um grafo e associar cores distintas a cada vértice.

```python
def colore_vertices(G):
    paleta = {0:'blue', 1:"Green", 2:"red", 3:"orange", 4:"violet"}
    color_map = []
    # Determina o número cromático do Grafo, ou seja, a menor quantidade de cores que colore o grafo conforme as restrições
    cor_vertices = nx.coloring.greedy_color(G, strategy="largest_first")
    nr_cromatico = max(cor_vertices.values())+1
    print("Número mínimo de Cores:", nr_cromatico)
    
    # define a cor de cada vértice
    for node in G:
        color_map.append(paleta[cor_vertices[node]])
    return color_map
```

```python
# Definição do grafo
G = nx.Graph()

V = ["A", "B", "C", "D", "E", "F", "G"]
E = [("A", "B"), ("A", "C"), ("A", "D"), ("A", "E"), ("A", "F"), ("A", "G"),
    ("B", "F"), ("B", "G"), ("C", "F"), ("D", "F"), ("D", "G")]
G.add_nodes_from(V)
G.add_edges_from(E)

nx.draw_circular(G, with_labels=True, font_weight='bold', font_color="w", node_color='b')
```

```python
# Mostra o grafo colorido
color_map = colore_vertices(G)
nx.draw_circular(G, with_labels=True, font_weight='bold', font_color="w", node_color=color_map)
```
